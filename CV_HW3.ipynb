````python
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

print("TensorFlow:", tf.__version__)
print("GPU available:", tf.config.list_physical_devices('GPU'))

# ========== 2. ЗАГРУЗКА ДАННЫХ ==========
IMG_SIZE = 224
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

(train_ds, val_ds, test_ds), ds_info = tfds.load(
    "tf_flowers",
    split=["train[:70%]", "train[70%:85%]", "train[85%:]"],
    as_supervised=True,
    with_info=True,
)

NUM_CLASSES = ds_info.features["label"].num_classes
print("Classes:", ds_info.features["label"].names, "| №:", NUM_CLASSES)

# Функция для аугментации данных
def augment(image, label):
    # Случайное горизонтальное отражение
    image = tf.image.random_flip_left_right(image)
    # Случайное изменение яркости
    image = tf.image.random_brightness(image, 0.1)
    # Случайное изменение контраста
    image = tf.image.random_contrast(image, 0.8, 1.2)
    # Случайное изменение насыщенности
    image = tf.image.random_saturation(image, 0.8, 1.2)
    # Случайный поворот
    image = tf.image.rot90(image, tf.random.uniform(shape=[], minval=0, maxval=4, dtype=tf.int32))
    return image, label

def preprocess(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# Применение аугментации только к обучающему набору
train_ds = (train_ds
            .map(augment, num_parallel_calls=AUTOTUNE)
            .map(preprocess, num_parallel_calls=AUTOTUNE)
            .shuffle(1024)
            .batch(BATCH_SIZE)
            .prefetch(AUTOTUNE))

val_ds = (val_ds
          .map(preprocess, num_parallel_calls=AUTOTUNE)
          .batch(BATCH_SIZE)
          .prefetch(AUTOTUNE))

test_ds = (test_ds
           .map(preprocess, num_parallel_calls=AUTOTUNE)
           .batch(BATCH_SIZE)
           .prefetch(AUTOTUNE))

# Визуализация примеров аугментированных изображений
plt.figure(figsize=(12, 8))
for i, (image, label) in enumerate(train_ds.unbatch().take(9)):
    plt.subplot(3, 3, i+1)
    plt.imshow(image)
    plt.title(ds_info.features["label"].names[label])
    plt.axis('off')
plt.suptitle("Примеры аугментированных изображений", fontsize=16)
plt.show()

# ========== 3. ЗАГРУЗКА И НАСТРОЙКА МОДЕЛЕЙ ==========
# Создаем несколько моделей для сравнения

# 1. ResNet50V2
def create_resnet_model():
    base_model = tf.keras.applications.ResNet50V2(
        weights="imagenet",
        include_top=False,
        input_shape=(IMG_SIZE, IMG_SIZE, 3)
    )
    base_model.trainable = False

    inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base_model(inputs, training=False)
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dropout(0.3)(x)  # Увеличен dropout для лучшей регуляризации
    outputs = tf.keras.layers.Dense(NUM_CLASSES, activation='softmax',
                                   kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)  # Добавлена L2 регуляризация

    model = tf.keras.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model, base_model, "ResNet50V2"

# 2. InceptionV3
def create_inception_model():
    base_model = tf.keras.applications.InceptionV3(
        weights="imagenet",
        include_top=False,
        input_shape=(IMG_SIZE, IMG_SIZE, 3)
    )
    base_model.trainable = False

    inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base_model(inputs, training=False)
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    outputs = tf.keras.layers.Dense(NUM_CLASSES, activation='softmax',
                                   kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)

    model = tf.keras.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model, base_model, "InceptionV3"

# Выбираем модель для обучения
model, base_model, model_name = create_resnet_model()
model.summary()

# ========== 4. ОБУЧЕНИЕ ==========
EPOCHS = 10
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint(
    f"{model_name}_flowers.h5",
    save_best_only=True,
    monitor="val_accuracy",
    mode="max"
)

# Добавляем раннюю остановку
early_stopping_cb = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=3,
    restore_best_weights=True
)

# Добавляем снижение learning rate при плато
reduce_lr_cb = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.2,
    patience=2,
    min_lr=1e-6
)

history = model.fit(
    train_ds,
    epochs=EPOCHS,
    validation_data=val_ds,
    callbacks=[checkpoint_cb, early_stopping_cb, reduce_lr_cb]
)

# Визуализация процесса обучения
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='validation')
plt.title('Точность')
plt.xlabel('Эпоха')
plt.ylabel('Точность')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='validation')
plt.title('Функция потерь')
plt.xlabel('Эпоха')
plt.ylabel('Потери')
plt.legend()
plt.tight_layout()
plt.show()

# Вторая фаза дообучки
UNFREEZE = 20
base_model.trainable = True
for layer in base_model.layers[:-UNFREEZE]:
    layer.trainable = False
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
history_fine = model.fit(
    train_ds,
    epochs=EPOCHS // 2,
    validation_data=val_ds,
    callbacks=[checkpoint_cb, early_stopping_cb, reduce_lr_cb]
)

# ========== 5. ОЦЕНКА ==========
model.load_weights(f"{model_name}_flowers.h5")  # Лучшая версия
y_true, y_pred = [], []
for images, labels in test_ds:
    preds = model.predict(images, verbose=0)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))
y_true = np.array(y_true)
y_pred = np.array(y_pred)

print("\n=== Classification report ===")
target_names = ds_info.features["label"].names
print(classification_report(y_true, y_pred, target_names=target_names, digits=4))

# Визуализация матрицы ошибок
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Предсказанные классы')
plt.ylabel('Истинные классы')
plt.title('Матрица ошибок')
plt.show()

# ========== 6. ВИЗУАЛИЗАЦИЯ ПРОГНОЗОВ ==========
correct_idx = np.where(y_true == y_pred)[0]
wrong_idx = np.where(y_true != y_pred)[0]

def plot_examples(idxs, title):
    plt.figure(figsize=(15, 3))
    for i, idx in enumerate(idxs[:5]):
        image, label = list(test_ds.unbatch().skip(idx).take(1))[0]
        pred = y_pred[idx]

        # Получаем вероятности для всех классов
        probs = model.predict(tf.expand_dims(image, 0), verbose=0)[0]
        pred_prob = probs[pred]

        plt.subplot(1, 5, i+1)
        plt.imshow(image)
        plt.axis('off')
        plt.title(f"T: {target_names[label]}\nP: {target_names[pred]} ({pred_prob:.2f})")
    plt.suptitle(title, fontsize=16)
    plt.show()

plot_examples(correct_idx, "Правильные предсказания")
plot_examples(wrong_idx, "Ошибочные предсказания")

# Анализ ошибок
def analyze_errors():
    error_analysis = {}
    for idx in wrong_idx:
        true_label = y_true[idx]
        pred_label = y_pred[idx]
        true_class = target_names[true_label]
        pred_class = target_names[pred_label]

        key = f"{true_class} → {pred_class}"
        if key not in error_analysis:
            error_analysis[key] = 0
        error_analysis[key] += 1

    # Сортировка по количеству ошибок
    sorted_errors = sorted(error_analysis.items(), key=lambda x: x[1], reverse=True)

    # Вывод наиболее частых ошибок
    print("Наиболее частые ошибки:")
    for error, count in sorted_errors[:5]:
        print(f"{error}: {count} случаев")

analyze_errors()
````

### 1. Данные и препроцессинг

* **Набор**: `tf_flowers` (3670 изобр., 5 классов).
* **Сплит**: 70 % train / 15 % val / 15 % test (автоматически через `tfds`).
* **Аугментации**: случайный флип, контраст/яркость/насыщенность, поворот 90°×k.
* **Размер входа**: 224 × 224; нормализация в \[0;1].

### 2. Архитектура и обучение

| Этап                                            | Настройки                                 | val accuracy     | val loss      |
| ----------------------------------------------- | ----------------------------------------- | ---------------- | ------------- |
| **Замороженная ResNet50V2** (top-6 объектов)    | Adam 1e-4, 10 эпох, Dropout 0.3, L2 0.001 | ↑ от 0.53 → 0.81 | ↓ 1.19 → 0.52 |
| **Fine-tuning** (раскрыты **20** верхних слоёв) | Adam 1e-5, 5 эпох                         | ↑ 0.81 → 0.87    | ↓ 0.52 → 0.37 |

* Переход к меньшему LR после разморозки стабилизировал обучение, не дав «скачка» потерь.
* Callbacks: `ModelCheckpoint`, `EarlyStopping(patience=3)`, `ReduceLROnPlateau`.

### 4. Анализ ошибок

| Переход (истина → предсказание) | Число случаев | Возможная причина                       |
| ------------------------------- | ------------- | --------------------------------------- |
| **roses → tulips**              | 16            | сходный цвет бутонов, узкие стебли      |
| sunflowers → tulips             | 8             | крупный жёлтый объект без видимых семян |

Снимки «bad-cases» подтверждают: модель «путается», когда в кадре — малое число лепестков или объект частично закрыт.

````python
IMG_DIR = DATA_ROOT/'PNGImages'
MASK_DIR = DATA_ROOT/'PedMasks'
YOLO_DIR = Path('/content/penn_yolo')
for sub in ['images/train','images/val','labels/train','labels/val']:
    (YOLO_DIR/sub).mkdir(parents=True, exist_ok=True)

def mask2yolo(idx_png):
    mask = cv2.imread(str(idx_png), cv2.IMREAD_GRAYSCALE)
    h,w = mask.shape
    obj_ids = np.unique(mask)[1:]                  # 0 — фон
    boxes = []
    for oid in obj_ids:
        ys, xs = np.where(mask == oid)
        xmin,xmax,ymin,ymax = xs.min(), xs.max(), ys.min(), ys.max()
        xc, yc = ((xmin+xmax)/2)/w, ((ymin+ymax)/2)/h
        bw, bh = (xmax-xmin)/w, (ymax-ymin)/h
        boxes.append(f"0 {xc} {yc} {bw} {bh}")
    return boxes

all_imgs = sorted(list(IMG_DIR.glob('*.png')))
random.seed(42); random.shuffle(all_imgs)
split = int(0.8*len(all_imgs))
train_imgs, val_imgs = all_imgs[:split], all_imgs[split:]

def convert_split(img_list, split_name):
    for img_path in tqdm.tqdm(img_list, desc=f'→ {split_name}'):
        lbl_path = YOLO_DIR/f"labels/{split_name}/{img_path.stem}.txt"
        with open(lbl_path,'w') as f:
            f.write('\n'.join(mask2yolo(MASK_DIR/f"{img_path.stem}_mask.png")))
        shutil.copy(img_path, YOLO_DIR/f"images/{split_name}/{img_path.name}")

import tqdm
convert_split(train_imgs,'train')
convert_split(val_imgs,'val')
print('Готово! YOLO-файлов:', len(list((YOLO_DIR/'labels/train').glob('*.txt'))))
````

````python
data_yaml = {
    'path': str(YOLO_DIR),
    'train': 'images/train',
    'val'  : 'images/val',
    'nc'   : 1,
    'names': ['pedestrian']
}
with open(YOLO_DIR/'pennped.yaml','w') as f: yaml.dump(data_yaml, f)
print((YOLO_DIR/'pennped.yaml').read_text())
````

````python
model = YOLO('yolov8n.pt')          # nano ≈ 3.2 M парам.
results = model.train(
    data  = YOLO_DIR/'pennped.yaml',
    epochs=50,
    imgsz =640,
    batch =16,
    device=0 if torch.cuda.is_available() else 'cpu',
    patience=10,
    name='pennped_yolov8n')
````
