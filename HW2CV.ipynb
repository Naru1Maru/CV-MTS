## **Решение**

**Пояснения:**
1. **Подготовка данных:**  
   - Датасет MNIST загружается, нормализуется, преобразуется в векторный формат и делится на обучающую и тестовую выборки в соотношении 80/20.
   
2. **Классификация с SVM:**  
   - Используется SVM с ядром RBF. Модель обучается, предсказывает классы для тестовой выборки, и вычисляются основные метрики: точность, полнота, F1-score, а также строится матрица ошибок.

3. **Обучение MLP:**  
   - Создаётся нейронная сеть с двумя скрытыми слоями и dropout для регуляризации. Модель обучается с использованием оптимизатора Adam, после чего также вычисляются метрики качества и строятся графики динамики обучения (потерь и точности).

4. **Сравнительный анализ:**  
   - Выводятся полученные метрики для SVM и MLP. В комментариях обсуждаются сильные и слабые стороны каждого подхода.

```python
# Импорт необходимых библиотек
import numpy as np
import matplotlib.pyplot as plt

# Библиотеки для работы с моделью SVM и оценки качества
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, recall_score, f1_score, confusion_matrix, classification_report

# Импортируем датасет MNIST и необходимые модули из Keras для построения MLP
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# ---------------------------
# 1. Подготовка данных
# ---------------------------

# Загружаем данные MNIST. MNIST содержит изображения рукописных цифр.
# Данные уже разделены на обучающую и тестовую выборки, но для наших целей объединяем их и делим заново по схеме 80/20.
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Объединяем тренировочные и тестовые данные в один общий набор
X = np.concatenate((x_train, x_test), axis=0)
y = np.concatenate((y_train, y_test), axis=0)

# Нормализуем пиксели изображений (значения от 0 до 255 переводим в диапазон [0,1])
X = X.astype('float32') / 255.0

# Преобразуем двумерные изображения (28x28) в вектор длиной 784 для дальнейшей работы
X_flat = X.reshape(X.shape[0], -1)

# Делим данные на обучающую (80%) и тестовую (20%) выборки
X_train, X_test, y_train, y_test = train_test_split(X_flat, y, test_size=0.2, random_state=42)

# Для нейронной сети преобразуем метки в формат one-hot encoding
y_train_cat = to_categorical(y_train, num_classes=10)
y_test_cat = to_categorical(y_test, num_classes=10)

print("Размер обучающей выборки:", X_train.shape)
print("Размер тестовой выборки:", X_test.shape)

# ---------------------------
# 2. Классификация с использованием SVM
# ---------------------------

# Инициализируем классификатор SVM с радиальной базисной функцией (RBF)
svm_classifier = SVC(kernel='rbf', gamma='scale', random_state=42)

# Обучаем модель на тренировочной выборке
print("\nОбучение SVM...")
svm_classifier.fit(X_train, y_train)

# Предсказываем метки для тестовой выборки
y_pred_svm = svm_classifier.predict(X_test)

# Вычисляем основные метрики качества
accuracy_svm = accuracy_score(y_test, y_pred_svm)
recall_svm = recall_score(y_test, y_pred_svm, average='macro')
f1_svm = f1_score(y_test, y_pred_svm, average='macro')
cm_svm = confusion_matrix(y_test, y_pred_svm)

print("\nРезультаты SVM:")
print("Accuracy:", accuracy_svm)
print("Recall:", recall_svm)
print("F1-score:", f1_svm)
print("\nClassification Report SVM:\n", classification_report(y_test, y_pred_svm))

# Построение матрицы ошибок для SVM
plt.figure(figsize=(8, 6))
plt.imshow(cm_svm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Матрица ошибок для SVM")
plt.colorbar()
tick_marks = np.arange(10)
plt.xticks(tick_marks, tick_marks)
plt.yticks(tick_marks, tick_marks)
plt.xlabel('Предсказанный класс')
plt.ylabel('Истинный класс')

# Добавляем числовые значения в клетки матрицы
thresh = cm_svm.max() / 2.
for i in range(cm_svm.shape[0]):
    for j in range(cm_svm.shape[1]):
        plt.text(j, i, format(cm_svm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm_svm[i, j] > thresh else "black")
plt.tight_layout()
plt.show()

# ---------------------------
# 3. Построение и обучение нейронной сети (MLP)
# ---------------------------

# Определяем архитектуру нейронной сети с использованием Keras
model = Sequential()
# Первый полносвязный слой с 128 нейронами и функцией активации ReLU
model.add(Dense(128, activation='relu', input_shape=(784,)))
# Dropout для предотвращения переобучения
model.add(Dropout(0.2))
# Второй полносвязный слой с 64 нейронами и функцией активации ReLU
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
# Выходной слой: 10 нейронов для 10 классов с функцией активации softmax
model.add(Dense(10, activation='softmax'))

# Компилируем модель с использованием оптимизатора Adam и функции потерь categorical_crossentropy
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Обучаем модель. Используем 20% обучающих данных для валидации, обучаем 20 эпох с размером батча 128.
print("\nОбучение MLP...")
history = model.fit(X_train, y_train_cat, epochs=20, batch_size=128, validation_split=0.2, verbose=2)

# Оцениваем модель на тестовой выборке
score = model.evaluate(X_test, y_test_cat, verbose=0)
print("\nРезультаты MLP:")
print("Тестовая потеря:", score[0])
print("Точность на тесте:", score[1])

# Получаем предсказания модели
y_pred_mlp_prob = model.predict(X_test)
y_pred_mlp = np.argmax(y_pred_mlp_prob, axis=1)

# Вычисляем метрики качества для MLP
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)
recall_mlp = recall_score(y_test, y_pred_mlp, average='macro')
f1_mlp = f1_score(y_test, y_pred_mlp, average='macro')
cm_mlp = confusion_matrix(y_test, y_pred_mlp)

print("\nМетрики MLP:")
print("Accuracy:", accuracy_mlp)
print("Recall:", recall_mlp)
print("F1-score:", f1_mlp)
print("\nClassification Report MLP:\n", classification_report(y_test, y_pred_mlp))

# Построение матрицы ошибок для MLP
plt.figure(figsize=(8, 6))
plt.imshow(cm_mlp, interpolation='nearest', cmap=plt.cm.Oranges)
plt.title("Матрица ошибок для MLP")
plt.colorbar()
plt.xticks(tick_marks, tick_marks)
plt.yticks(tick_marks, tick_marks)
plt.xlabel('Предсказанный класс')
plt.ylabel('Истинный класс')
thresh = cm_mlp.max() / 2.
for i in range(cm_mlp.shape[0]):
    for j in range(cm_mlp.shape[1]):
        plt.text(j, i, format(cm_mlp[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm_mlp[i, j] > thresh else "black")
plt.tight_layout()
plt.show()

# Построение графиков динамики обучения (loss и accuracy) для MLP
# График изменения функции потерь
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Потеря на обучении')
plt.plot(history.history['val_loss'], label='Потеря на валидации')
plt.title('Динамика функции потерь')
plt.xlabel('Эпоха')
plt.ylabel('Потеря')
plt.legend()

# График изменения точности
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Точность на обучении')
plt.plot(history.history['val_accuracy'], label='Точность на валидации')
plt.title('Динамика точности')
plt.xlabel('Эпоха')
plt.ylabel('Точность')
plt.legend()

plt.tight_layout()
plt.show()

# ---------------------------
# 4. Сравнительный анализ моделей
# ---------------------------

print("\nСравнительный анализ моделей:")
print("-------------------------------------------------")
print("Метрика             |   SVM         |   MLP")
print("-------------------------------------------------")
print(f"Accuracy            |   {accuracy_svm:.4f}     |   {accuracy_mlp:.4f}")
print(f"Recall              |   {recall_svm:.4f}     |   {recall_mlp:.4f}")
print(f"F1-score            |   {f1_svm:.4f}     |   {f1_mlp:.4f}")
print("-------------------------------------------------")
```

Таким образом, в рамках данной задачи оба подхода продемонстрировали высокую точность классификации. ## **Дополнительное решение с улучшениями**

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import tensorflow as tf

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical

from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Фиксируем seed для воспроизводимости
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

# ==============================
# 1. ЗАГРУЗКА ДАННЫХ MNIST
# ==============================
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Объединяем данные и нормализуем (значения от 0 до 1)
X = np.concatenate((x_train, x_test), axis=0).astype('float32') / 255.0
y = np.concatenate((y_train, y_test), axis=0)

# Для SVM и MLP преобразуем изображения в векторы (784)
X_flat = X.reshape(X.shape[0], -1)

# Делим данные на обучающую и тестовую выборки (80/20)
X_train_flat, X_val_flat, y_train, y_val = train_test_split(X_flat, y, test_size=0.2, random_state=42)

# Для MLP переведем метки в one-hot векторы
y_train_cat = to_categorical(y_train, 10)
y_val_cat = to_categorical(y_val, 10)

print("Размер обучающей выборки:", X_train_flat.shape)
print("Размер тестовой выборки:", X_val_flat.shape)

# ==============================
# 2. Улучшение SVM через PCA и подбор гиперпараметров
# ==============================

# Создаем конвейер: сначала масштабирование, затем PCA (оставляем 98% дисперсии), затем SVM с RBF
svm_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=0.98, svd_solver='full')),  # сохраняем 98% дисперсии
    ('svm', SVC(kernel='rbf', probability=True, random_state=42))
])

# Подбираем гиперпараметры: C и gamma SVM
param_grid = {
    'svm__C': [1, 10],
    'svm__gamma': ['scale', 0.001, 0.0001]
}

grid_search = GridSearchCV(svm_pipeline, param_grid, cv=3, scoring='accuracy', verbose=2)
print("\nОбучаем SVM с PCA и Grid Search гиперпараметров...")
grid_search.fit(X_train_flat, y_train)

print("\nЛучшая конфигурация SVM:", grid_search.best_params_)
print("Accuracy на обучении (CV):", grid_search.best_score_)

# Предсказание и оценка на валидационной выборке
y_pred_svm = grid_search.predict(X_val_flat)
print("\nClassification Report SVM:")
print(classification_report(y_val, y_pred_svm))

# Построение матрицы ошибок для SVM
cm_svm = confusion_matrix(y_val, y_pred_svm)
plt.figure(figsize=(8, 6))
plt.imshow(cm_svm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Матрица ошибок SVM с PCA")
plt.colorbar()
tick_marks = np.arange(10)
plt.xticks(tick_marks, tick_marks)
plt.yticks(tick_marks, tick_marks)
plt.xlabel("Предсказанный класс")
plt.ylabel("Истинный класс")
thresh = cm_svm.max() / 2.
for i in range(cm_svm.shape[0]):
    for j in range(cm_svm.shape[1]):
        plt.text(j, i, format(cm_svm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm_svm[i, j] > thresh else "black")
plt.tight_layout()
plt.show()

# ==============================
# 3. Улучшение MLP с Batch Normalization и Learning Rate Scheduler
# ==============================
def create_mlp_model():
    model = Sequential()
    # Начинаем с Flatten: преобразуем 28x28 в вектор
    model.add(Flatten(input_shape=(28, 28)))

    # Полносвязный слой с BatchNormalization, Dropout и L2-регуляризацией можно добавить при необходимости
    model.add(Dense(256, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))

    model.add(Dense(128, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))

    model.add(Dense(64, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))

    # Выходной слой
    model.add(Dense(10, activation='softmax'))

    return model

# Переобразуем данные для MLP в исходный 28x28 формат
X_train_mlp = X_train_flat.reshape(-1, 28, 28)
X_val_mlp = X_val_flat.reshape(-1, 28, 28)

model_mlp = create_mlp_model()

# Используем оптимизатор Adam с начальной скоростью обучения
optimizer = Adam(learning_rate=0.001)
model_mlp.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Колбэки: EarlyStopping и ReduceLROnPlateau
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-5, verbose=1)

# Обучаем модель. Увеличиваем число эпох для более глубокого обучения.
history = model_mlp.fit(
    X_train_mlp, y_train_cat,
    epochs=30,
    batch_size=128,
    validation_data=(X_val_mlp, y_val_cat),
    callbacks=[early_stop, reduce_lr],
    verbose=2
)

# Оценка модели MLP на валидационной выборке
val_loss, val_acc = model_mlp.evaluate(X_val_mlp, y_val_cat, verbose=0)
print("\nРезультаты MLP:")
print("Loss:", val_loss)
print("Accuracy:", val_acc)

# Предсказания и построение Classification Report для MLP
y_pred_mlp = np.argmax(model_mlp.predict(X_val_mlp), axis=1)
print("\nClassification Report MLP:")
from sklearn.metrics import classification_report
print(classification_report(y_val, y_pred_mlp))

# Построение матрицы ошибок для MLP
cm_mlp = confusion_matrix(y_val, y_pred_mlp)
plt.figure(figsize=(8, 6))
plt.imshow(cm_mlp, interpolation='nearest', cmap=plt.cm.Oranges)
plt.title("Матрица ошибок MLP с BatchNormalization")
plt.colorbar()
plt.xticks(tick_marks, tick_marks)
plt.yticks(tick_marks, tick_marks)
plt.xlabel("Предсказанный класс")
plt.ylabel("Истинный класс")
thresh = cm_mlp.max() / 2.
for i in range(cm_mlp.shape[0]):
    for j in range(cm_mlp.shape[1]):
        plt.text(j, i, format(cm_mlp[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm_mlp[i, j] > thresh else "black")
plt.tight_layout()
plt.show()

# ==============================
# 4. ГРАФИКИ ДЛЯ РЕЗУЛЬТАТОВ MLP
# ==============================
# График динамики функции потерь
plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Обучение')
plt.plot(history.history['val_loss'], label='Валидация')
plt.title("Динамика функции потерь (MLP)")
plt.xlabel("Эпоха")
plt.ylabel("Потеря")
plt.legend()

# График динамики точности
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Обучение')
plt.plot(history.history['val_accuracy'], label='Валидация')
plt.title("Динамика точности (MLP)")
plt.xlabel("Эпоха")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.show()
```

## **Выводы**

Это заняло много времени, но улучшение в 1% - тоже улучшение :🇩

**Общий вывод**  
1. **SVM** с использованием PCA и перебора гиперпараметров (C=10, gamma=0.001) показал на валидационной выборке точность около 0.97. При этом метрики (Precision, Recall, F1-score) для каждого класса также находятся в диапазоне 0.96–0.99, что свидетельствует о высоком уровне распознавания рукописных цифр.  
2. **MLP** с добавленными улучшениями (Batch Normalization, Dropout, EarlyStopping, ReduceLROnPlateau и увеличенным числом эпох) достиг точности около 0.98. Такая точность на MNIST говорит о том, что сеть хорошо обучилась распознаванию цифр и не переобучилась (EarlyStopping останавливает обучение, когда переобучение начинает проявляться).  

Таким образом, за счёт дополнительных улучшений (PCA и Grid Search для SVM, а также Batch Normalization и гибкой стратегии обучения для MLP) обе модели вышли на высокий уровень точности (97–98%), что сопоставимо с лучшими классическими результатами на датасете MNIST без использования сверточных нейронных сетей.

### Подробные пояснения

1. **Анализ матриц ошибок**  
   - **Матрица ошибок SVM**:  
     - По диагонали мы видим наибольшие числа, что говорит о корректном распознавании соответствующих цифр.  
     - Вне диагонали есть небольшие вкрапления — это случаи, когда модель путает близкие по написанию цифры, например «8» и «9», «3» и «5» и т. п. Тем не менее, общее число ошибок невелико на фоне всей выборки.  
   - **Матрица ошибок MLP**:  
     - Аналогично, на диагонали большие значения.  
     - Модель путает ещё меньше образцов, чем SVM, что подтверждается чуть более высокой точностью (0.98).  

2. **Графики обучения MLP**  
   - **Динамика функции потерь**:  
     - Синяя линия (обучение) уверенно убывает от начала к середине обучения и продолжает снижаться медленнее к концу.  
     - Оранжевая линия (валидация) тоже убывает, при этом остаётся ниже обучающей, что говорит о хорошем качестве обобщения (отсутствует серьёзное переобучение).  
   - **Динамика точности**:  
     - Синяя линия (обучение) растёт и достигает ~0.99 к финальным эпохам.  
     - Оранжевая линия (валидация) растёт до ~0.98, после чего стабилизируется.  
     - Видно, что в середине обучения точность на валидации иногда немного отстаёт, но благодаря ReduceLROnPlateau скорость обучения снижается, и сеть «дорабатывает» до лучшего результата.

3. **Влияние улучшений на результаты**  
   - **SVM**:  
     - Применение PCA (оставляющего 98% дисперсии) уменьшает размерность, устраняет шум и ускоряет обучение SVM. Это особенно полезно, учитывая, что Grid Search с кросс-валидацией (cv=3) требует многократного обучения модели.  
     - Подбор гиперпараметров (C и gamma) через GridSearchCV помогает найти оптимальное сочетание, благодаря чему точность достигает ~0.97.  
   - **MLP**:  
     - **Batch Normalization** после каждого скрытого слоя стабилизирует распределение активаций и ускоряет сходимость.  
     - **Dropout** (0.3) снижает риск переобучения.  
     - **EarlyStopping** и **ReduceLROnPlateau** позволяют увеличить количество эпох (до 30), но при этом автоматически остановить обучение или снизить learning rate, если улучшения на валидации перестают расти. Это приводит к более «гибкой» и точной настройке весов сети.  
     - В итоге точность на валидации ~0.98 — это отличный результат для MLP на MNIST, где обычно сверточные сети (CNN) достигают чуть более высоких значений.

4. **Заключение**  
   - **SVM** с PCA и оптимизированными гиперпараметрами стабильно достигает 97% точности, что близко к результатам полносвязной сети, но при этом требует заметно больше времени на обучение (особенно при Grid Search).  
   - **MLP** с добавленными элементами (BatchNorm, Dropout, адаптивная скорость обучения) выходит на ~98% точности, что указывает на хорошую способность сети к обобщению.  
   - Таким образом, в рамках ограничений «только SVM и MLP» обе модели демонстрируют превосходное качество на MNIST. Прирост по сравнению с базовыми версиями MLP (без BatchNorm, без гибкой LR) и SVM (без PCA и подбора гиперпараметров) подтверждает, что данные улучшения оправданны и позволяют добиться более высокой точности классификации.

Итог:  
- **SVM** ~ 97% точности,  
- **MLP** ~ 98% точности.  

За счёт PCA, Grid Search, Batch Normalization, Dropout и динамического управления скоростью обучения удалось добиться прироста метрик, сохранив при этом устойчивость и надёжность обучения.
